```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
```

```python
sf=pd.DataFrame(diabetes['data'])
sf
```

```python
diabetes_labels=diabetes.target
diabetes.target.shape
```

```python
diabetes_labels
```

```python
labels=np.reshape(diabetes_labels,(442,1))
labels.shape
labels
```

```python
sf=np.concatenate([diabetes.data,labels],axis=1)
sf.shape
sf=pd.DataFrame(sf)
sf
```

```python
features=diabetes.feature_names
features
```

```python
feature_labels=np.append(features,'labels')
feature_labels
```

```python
sf.columns=feature_labels
sf
```

```python
import seaborn as sns
sns.jointplot(sf['age'],sf['labels'],kind='reg')
```

```python
sns.pairplot(sf,hue='labels',size=2.4)
```

```python
sns.scatterplot(sf['age'],sf['sex'],hue=sf['labels'])
```

```python
x=sf.drop('labels',axis=1)
x
```

```python
y=sf['labels']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn import svm
classifier=svm.SVC(gamma='auto',C=2)
```

```python
classifier.fit(x_train,y_train)
```

```python
predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```
