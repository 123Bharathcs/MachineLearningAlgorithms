as we can see that our dataset is completely unbalanced anf if we do modeling our model might get biased and give the the proper accuracy with these imbalanced dtaset. So first we will try to balance the dataset by using under sampling method

```python
#splitting our dataset into independent and dependent features

x=data.drop(['Class'],axis=1)
x.head()
x.shape

y=data['Class']
y.head()
y.shape
```

```python
from imblearn import under_sampling
from imblearn.under_sampling import NearMiss
```

```python
nm=NearMiss()
x1,y1=nm.fit_sample(x,y)
x1.shape
y1.shape
```

```python
plt.figure(figsize=(8,6))
sns.countplot(y1,hue=y1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.20,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_predict)
print("confusionmatrix:",cm)
print("accuracy_score:",accuracy_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
